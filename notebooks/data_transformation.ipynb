```python
df.drop(columns=['Vehicle Images'], inplace=True)

df['Date'] = pd.to_datetime(df['Date'])

df['TAT_Applicable'] = (df['Booking_Status'] == 'Success').astype(int)
df['Is_Rated_Ride'] = (df['Booking_Status'] == 'Success').astype(int)

df['Payment_Method'] = df['Payment_Method'].fillna('Not Applicable')
df['Incomplete_Rides'] = df['Incomplete_Rides'].fillna('Not Applicable')

df['Incomplete_Rides_Reason'] = df['Incomplete_Rides_Reason'].fillna('Not Applicable')
df.loc[df['Booking_Status'] == 'Success', 'Incomplete_Rides_Reason'] = 'Ride Completed'

df['Canceled_Rides_by_Customer'] = df['Canceled_Rides_by_Customer'].fillna('Not Applicable')
df['Canceled_Rides_by_Driver'] = df['Canceled_Rides_by_Driver'].fillna('Not Applicable')
```

```python
df.drop(columns=['Vehicle Images'],inplace=True) 
```

```python
df['Date']=pd.to_datetime(df['Date'])
```

```python
df['Is_Rated_Ride'] = df['Customer_Rating'].notna().astype(int) # rated or not filter col

df['Driver_Ratings'] = df['Driver_Ratings'].fillna(-1)
df['Customer_Rating'] = df['Customer_Rating'].fillna(-1)
```

```python
df['Payment_Method'] = df['Payment_Method'].fillna('Not Applicable') # fill na payment method 
```

```python
df['TAT_Applicable'] = df['C_TAT'].notna().astype(int) #tat applicable or not , then fill tat with -1 whre not applicable

df['C_TAT']=df['C_TAT'].fillna(-1)
df['V_TAT']=df['V_TAT'].fillna(-1)
```

```python
df['Incomplete_Rides'] = (df['Incomplete_Rides'].fillna('Not Applicable')) # nan are not booked rides
```

```python
# filling Incomplete_Rides_Reason nan

df.loc[df['Booking_Status'] == 'Success','Incomplete_Rides_Reason'] = df.loc[df['Booking_Status'] == 'Success','Incomplete_Rides_Reason'].fillna('Ride Complete')

df.loc[df['Booking_Status']=='Canceled by Customer', 'Incomplete_Rides_Reason'] = df.loc[df['Booking_Status']=='Canceled by Customer', 'Incomplete_Rides_Reason'].fillna('Canceled by Customer')

df.loc[df['Booking_Status']=='Canceled by Driver', 'Incomplete_Rides_Reason'] = df.loc[df['Booking_Status']=='Canceled by Driver', 'Incomplete_Rides_Reason'].fillna('Canceled by Driver')

df.loc[df['Booking_Status']=='Driver Not Found', 'Incomplete_Rides_Reason'] = df.loc[df['Booking_Status']=='Driver Not Found', 'Incomplete_Rides_Reason'].fillna('Driver Not Found')
```

```python
# filling canceled by customer nan

df.loc[df['Booking_Status']=='Success','Canceled_Rides_by_Customer'] = df.loc[df['Booking_Status']=='Success','Canceled_Rides_by_Customer'].fillna('Ride Complete')

df.loc[df['Booking_Status']=='Canceled by Driver','Canceled_Rides_by_Customer'] = df.loc[df['Booking_Status']=='Canceled by Driver','Canceled_Rides_by_Customer'].fillna('Canceled by Driver')

df.loc[df['Booking_Status']=='Driver Not Found','Canceled_Rides_by_Customer'] = df.loc[df['Booking_Status']=='Driver Not Found','Canceled_Rides_by_Customer'].fillna('Driver Not Found')
```

```python
# filling canceled by driver nan

df.loc[df['Booking_Status']=='Success','Canceled_Rides_by_Driver'] = df.loc[df['Booking_Status']=='Success','Canceled_Rides_by_Driver'].fillna('Ride Complete')

df.loc[df['Booking_Status']=='Canceled by Customer','Canceled_Rides_by_Driver'] = df.loc[df['Booking_Status']=='Canceled by Customer','Canceled_Rides_by_Driver'].fillna('Canceled by Customer')

df.loc[df['Booking_Status']=='Driver Not Found','Canceled_Rides_by_Driver'] = df.loc[df['Booking_Status']=='Driver Not Found','Canceled_Rides_by_Driver'].fillna('Driver Not Found')
```

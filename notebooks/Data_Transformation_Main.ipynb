```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

from dotenv import load_dotenv
import os
load_dotenv()
```

```python
df = pd.read_csv(os.getenv('csv_path'))
```

```python
pd.set_option('display.max_columns',None)
```

```python
# dropping vechcal image col
df.drop(columns=['Vehicle Images'],axis=1, inplace=True) 
```

```python
# changing date to datetime
df['Date'] = pd.to_datetime(df['Date']) 
```

```python
# creating a col to check tat applicability and filling c_tat and v_tat
df['TAT_Applicable'] = (df['Booking_Status']=='Success').astype(int)

# creating a col to check if the ride is rated or not
df['Rated_Ride_Applicable'] = (df['Booking_Status']=='Success').astype(int)
```

```python
# payment not applicable for rides not booked
df['Payment_Method'] = df['Payment_Method'].fillna('Not Applicable')
```

```python
# filling Incomplete_Rides_Reason nan
# df.loc[df['Booking_Status'] == 'Success','Incomplete_Rides_Reason'] = df.loc[df['Booking_Status'] == 'Success','Incomplete_Rides_Reason'].fillna('Not Applicable')
# df.loc[df['Booking_Status']=='Canceled by Customer', 'Incomplete_Rides_Reason'] = df.loc[df['Booking_Status']=='Canceled by Customer', 'Incomplete_Rides_Reason'].fillna('Booking Canceled by Customer')
# df.loc[df['Booking_Status']=='Canceled by Driver', 'Incomplete_Rides_Reason'] = df.loc[df['Booking_Status']=='Canceled by Driver', 'Incomplete_Rides_Reason'].fillna('Booking Canceled by Driver')
# df.loc[df['Booking_Status']=='Driver Not Found', 'Incomplete_Rides_Reason'] = df.loc[df['Booking_Status']=='Driver Not Found', 'Incomplete_Rides_Reason'].fillna('Driver Not Found')
df['Incomplete_Rides_Reason'] = df['Incomplete_Rides_Reason'].fillna('Not Applicable')
```

```python
# filling incomplete rides
df['Incomplete_Rides'] = df['Incomplete_Rides'].fillna('Not Applicable')
```

```python
# filling canceled by customer nan
# df.loc[df['Booking_Status']=='Success','Canceled_Rides_by_Customer'] = df.loc[df['Booking_Status']=='Success','Canceled_Rides_by_Customer'].fillna('Ride Successful')
# df.loc[df['Booking_Status']=='Canceled by Driver','Canceled_Rides_by_Customer'] = df.loc[df['Booking_Status']=='Canceled by Driver','Canceled_Rides_by_Customer'].fillna('Canceled by Driver')
# df.loc[df['Booking_Status']=='Driver Not Found','Canceled_Rides_by_Customer'] = df.loc[df['Booking_Status']=='Driver Not Found','Canceled_Rides_by_Customer'].fillna('Driver Not Found')
df['Canceled_Rides_by_Customer'] = df['Canceled_Rides_by_Customer'].fillna('Not Applicable')
```

```python
# filling canceled by driver nan
# df.loc[df['Booking_Status']=='Success','Canceled_Rides_by_Driver'] = df.loc[df['Booking_Status']=='Success','Canceled_Rides_by_Driver'].fillna('Ride Successful')
# df.loc[df['Booking_Status']=='Canceled by Customer','Canceled_Rides_by_Driver'] = df.loc[df['Booking_Status']=='Canceled by Customer','Canceled_Rides_by_Driver'].fillna('Canceled by Customer')
# df.loc[df['Booking_Status']=='Driver Not Found','Canceled_Rides_by_Driver'] = df.loc[df['Booking_Status']=='Driver Not Found','Canceled_Rides_by_Driver'].fillna('Driver Not Found')
df['Canceled_Rides_by_Driver'] = df['Canceled_Rides_by_Driver'].fillna('Not Applicable')
```
